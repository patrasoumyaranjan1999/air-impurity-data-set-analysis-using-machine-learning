```python
#divide the dataset into 2 parts
X=df.drop(columns=["RH"])
Y=df["RH"]
X
```

```python
Y
```

```python
#scale down the values of X by using StandardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
X
```

```python
#split the dataset into 2 parts
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

# create the model of LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
L=LinearRegression()
#train the model
L.fit(X_train,Y_train)
```

```python
#test the model
Y_pred_LR=L.predict(X_test)
Y_pred_LR
```

```python
Y_test.values
```

```python
#find the mean squared error(mse)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_pred_LR,Y_test)
mse
```

```python
from sklearn.metrics import r2_score
rse=r2_score(Y_pred_LR,Y_test)
rse
```


# implementation of KNN model

```python
from sklearn.neighbors import KNeighborsRegressor
K=KNeighborsRegressor(n_neighbors=5)
K.fit(X_train,Y_train)
```

```python
Y_pred_KNN=K.predict(X_test)
Y_pred_KNN
```

```python
Y_test.values
```

```python
#find the mse
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_pred_KNN,Y_test)
mse
```

```python
#find the rse
from sklearn.metrics import r2_score
rse=r2_score(Y_pred_KNN,Y_test)
rse
```

# implementatiin DECISSION TREE REGRESSOR

```python
from sklearn.tree import DecisionTreeRegressor
D=DecisionTreeRegressor()
D.fit(X_train,Y_train)
```

```python
Y_pred_Tree=D.predict(X_test)
Y_pred_Tree
```

```python
Y_test.values
```

```python
#find the mse value
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_pred_Tree,Y_test)
mse
```

```python
from sklearn.metrics import r2_score
rse=r2_score(Y_pred_Tree,Y_test)
rse
```
